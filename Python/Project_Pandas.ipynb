**TODO 05** - how much total sales,average sales,and standard deviation of sale your company make in 2017

```python
sale_2017 = df[df["Order Date"].dt.year == 2017]
sale_2017
total_sale = sale_2017["Sales"].sum()
avg_sale = sale_2017["Sales"].mean()
std_sale = sale_2017["Sales"].std()
print(f"total sale of your company in 2017 is: {total_sale}")
print(f"average sale of your company in 2017 is: {avg_sale.round(4)}")
print(f"standard deviation of sale your company in 2017 is: {std_sale.round(4)}")
```

**TODO 06** - which Segment has the highest profit in 2018

```python
profit_2018 = df[df["Order Date"].dt.year == 2018]
profit_2018
sum_profit = profit_2018.groupby("Segment")["Profit"].sum().reset_index()
sum_profit
highest_profit = sum_profit[sum_profit["Profit"] == sum_profit["Profit"].max()]
highest_profit
#max_profit = max_profit.max()
#max_profit
```

**TODO 07** - Which top 5 States have the least total sales between 15 april 2019 - 31 December 2019

```python
start_date = pd.to_datetime('2019-04-15')
end_date = pd.to_datetime('2019-12-31')
least_sale = df[(df["Order Date"] >= start_date) & (df["Order Date"] <= end_date)]
top5_state = least_sale.groupby("State")["Sales"].sum()#
top5_least = top5_state.sort_values().iloc[0:5].reset_index()
top5_least
```

**TODO 08** - What is the proportion of total sales (%) in West + Central in 2019 e.g. 25%

```python
total_sale_2019 = df[df["Order Date"].dt.year == 2019]
total_sale_2019
total_sale_2019["Percent_Sales"] = (total_sale_2019["Sales"]/total_sale_2019["Sales"].sum())*100
total_sale_2019
total_sale_wc = total_sale_2019.groupby("Region")["Percent_Sales"].sum().reset_index()
total_sale_wc
proportion = total_sale_wc.query("Region == 'West' or Region == 'Central'").sum().reset_index()
proportion
#print(f"proportion of total sales (%) in West + Central in 2019 is: {proportion.round(2)}")
```

**TODO 09** - Find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

```python
df_2019_2020 = df[df["Order Date"].dt.year.isin([2019,2020])]
df_2019_2020
product = df_2019_2020.groupby("Product Name")[["Quantity","Sales"]].sum()
top10_product = product.sort_values("Quantity", ascending= False).head(10).reset_index()
top10_product
```
